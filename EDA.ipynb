```python
import warnings
warnings.simplefilter("ignore")

import konlpy
konlpy.__version__
```

```python
from konlpy.tag import *

hannanum = Hannanum()
kkma = Kkma()
okt = Okt()
```

```python
nm_horror = nv_df[(nv_df.genre == "공포")]
nm_horror.head(5)
```

```python
from nltk import Text

horror = Text(hannanum.nouns(" ".join(nm_horror["story"]).replace("\xa0","").replace("\xa0Q", "")), name="horror")
horror.plot(30)
plt.show()
```
